```python
# Importing opencv library
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

**Different Filtering techniques**

```python
# Setting random variables for plotting
x = np.linspace(1,20,100)
y = np.cos(x)

# Plotting the above points
plt.plot(x,y)
plt.title("Plot of f(x)")
plt.show()
```

**Adding some noise to the above plot**

```python
noise = np.random.randn(100)
noisy_signal = y+ 0.1*noise

# Re-plotting the above graph
plt.plot(x, noisy_signal)
plt.title("Noise signal")
plt.show()
```

```python
convolved_signal = np.convolve(noisy_signal, np.ones(5)/5)
plt.plot(x, convolved_signal[2:-2])
plt.show()
```

**Lets try apply this in real time !**

```python
# Loading another image from our directory
image = cv2.imread("./imgs/download.jfif")
images = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(image)
plt.show()
```

**Mean filter**

```python
plt.figure(figsize=(15,15))

for s, i in enumerate([3,7,11,25]):
    mean_filter = cv2.filter2D(images,-1, np.ones((i,i))/i**2)
    
    plt.title(f"Mean Filter {i}")
    plt.subplot(2,2,s+1)
    plt.imshow(mean_filter)

plt.show()
```

**Gaussian Blur**

```python
plt.figure(figsize=(15,15))

for s,i in enumerate([3,7,11,25]):
    
    img = cv2.GaussianBlur(image, (i,i), i/6)
    
    plt.title(f"Gaussian filter {i}")
    plt.subplot(2,2,s+1)
    plt.imshow(img)  

plt.show()
```

**Mean Filter**

```python
plt.figure(figsize= (7,7))
img = cv2.blur(image, (11,11))
plt.imshow(img)
plt.title("Mean Filter")
plt.show()
```

**Median Filter**

```python
plt.figure(figsize= (10,10))
img = cv2.medianBlur(image, 11)
plt.imshow(img)
plt.title("Median Filter")
plt.show()
```

**Gaussian Filter**

```python
plt.figure(figsize= (7,7))
img = cv2.GaussianBlur(image, (11,11), 11/6)
plt.imshow(img)
plt.title("Gaussian Filter")
plt.show()
```

**Face Detection**

```python
import cv2
 
# Create a CascadeClassifier Object
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
 
# Reading the image as it is
img = cv2.imread("imgs/cat.jpg")
if(img is not None):
    cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Reading the image as gray scale image
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# Search the co-ordintes of the image
faces = face_cascade.detectMultiScale(gray_img, scaleFactor = 1.05, minNeighbors=5)
for x,y,w,h in faces:
    img = cv2.rectangle(img, (x,y), (x+w,y+h),(0,255,0),3)
    
resized = cv2.resize(img, (int(img.shape[1]/7),int(img.shape[0]/7)))
cv2.imshow("Gray", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
